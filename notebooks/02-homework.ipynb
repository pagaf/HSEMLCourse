```python
df['Education'] = df['Education'].replace({
    'Basic': 0,
    'Graduation': 1,
    'Master': 2,
    '2n Cycle': 3,
    'PhD': 4
})
```

```python
df['Marital_Status'] = df['Marital_Status'].replace({
    'Single': 0,
    'Together': 1,
    'Married': 2,
    'Divorced': 3,
    'Widow': 4,
    'Alone': 5,
    'Absurd': 6,
    'YOLO': 7
})
```

```python
selected_columns=['Education','Marital_Status','Income','Kidhome','Teenhome','Recency','NumWebPurchases','NumStorePurchases','Age']
df_=df[selected_columns]
```

```python
df_.head()
```

```python
import seaborn as sns
sns.pairplot(df_)
plt.show()
```
